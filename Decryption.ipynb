```python
import cv2
import os
import string
import tkinter as tk
from tkinter import filedialog,simpledialog
from PIL import Image, ImageTk
from tkinter import messagebox
```

```python


def select_image():
    global image_path
    image_path = filedialog.askopenfilename(filetypes=[("Image Files", "*.png;*.jpg;*.jpeg;*.bmp;*.gif")])
    if image_path:
        
        img = Image.open(image_path)
        img = img.resize((250, 250))  # Resize the image to fit in the window
        img = ImageTk.PhotoImage(img)
        
        
        
        label.config(image=img)  # Update label with the new image
        label.image = img  # Store reference to avoid garbage collection
        label.pack(pady=10)
        
        
        text_label.config(text=f"Selected Image: {image_path.split('/')[-1]}", fg="black", font=("Arial", 12, "bold"))
        text_label.pack()


        
        
        
       # root.destroy()
        
        
        


root = tk.Tk()
root.title("Image Uploader")
root.geometry("600x600")
root.configure(bg="white")  



button = tk.Button(root, text="Click to Insert Image", font=("Arial", 12), 
                bg="gray", fg="white", width=20, height=2, bd=5, relief="flat",
                command=select_image)
button.pack(pady=50)





label = tk.Label(root, bg="white")
label.pack()

text_label = tk.Label(root, text="No image selected", bg="white", font=("Arial", 12))
text_label.pack()


root.mainloop()
```

```python
def custom_dialog(title, label_text):
    dialog = tk.Tk()
    dialog.title(title)
    dialog.geometry("300x200")
    dialog.configure(bg="lightgray")  # Background color

    tk.Label(dialog, text=label_text, font=("Comic Sans MS", 15), bg="lightgray").pack(pady=30)
    dialog.wait_window()  # Wait for user input

    
    
```
